#### Extracting Latitude and Longitude value using Geocoder 

Now , I will fetch the Latitude and Longitude of the neighbourhood using the geocoder and later I will create a  df with that .

Please note, for few Neighbourhood name , the required Data is not available . We can change the name of the neigbourhood , so that we can get latitude and longitude value . Such as 'Playter Estates-Danforth' with 'Danforth', 'Mount Olive-Silverstone-Jamestown' with 'Smithfield' etc . But in this project I have replaced the latitude and longitude of those neighbourhood with np.NAN . Later, I am going to remove those neighbourhood in my project .

```python
Lat_Long_array=[]
for neighbourhood in df["Neighbourhood_Name"]:
    address ="{} , Toronto".format(neighbourhood)
    print(neighbourhood)
    geolocator = Nominatim(user_agent="coursra_t")
    location = geolocator.geocode(address)
    if ( location is None):
        Lat_Long_array.append([neighbourhood,np.NAN,np.NAN])
    else:
        latitude = location.latitude
        longitude = location.longitude
        Lat_Long_array.append([neighbourhood,latitude,longitude])
```

Let me create a new Dataframe with these values and merge it with the original one 

```python
#creating a new df with latitude and longitude values
df_lat_long=pd.DataFrame(Lat_Long_array,columns=["Neighbourhood_Name","Latitude","Longitude"])

#merging this new df with the original df 
df=df.merge(df_lat_long,on="Neighbourhood_Name")

#atlast removing those neigbourhood which has NAN value in Latitude and longitude
df.dropna(axis=0,inplace=True)
```

In the next cell, a function is defined which will fetch venue details ( such as Venue name , Venue_latitude, venue_longitude and Venue_category ) for each and every neighboorhood. In tjis function I have set the radius is 500 m and to fetch at max 1000 venues for any neighbourhood . it can be increase or dcrease accordingly 

```python
def getNearbyVenues(names, latitudes, longitudes):
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            500, 
            1000)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood_Name', 
                  'Latitude', 
                  'Longitude', 
                  'Venue', 
                  'Venue_Latitude', 
                  'Venue_Longitude', 
                  'Venue_Category']
    
    return(nearby_venues)
```

```python
#Getting all the required data for those neighbourhood and put it in toronto_venues Datframe
toronto_venues = getNearbyVenues(names=df['Neighbourhood_Name'],
                                   latitudes=df['Latitude'],
                                   longitudes=df['Longitude']
                                  )
```

### Machine learning ( I have used K-means clustering)

Before using the k-means clustering in neigbourhoods let me explain what logic i have used to achive our goal .

Lets say, we have 5 neibourhoods such as N1, N2 ,N3 ,N4 and N5 and number of restaurant in those neighbourhoods are 10, 2 , 2 , 13, 4 respectively 

And After using clustering algorithm based on its features ( excluding the restaurent number) ,  lets say we have 2 cluster such as ( N1, N3 ,N4 ) and (N2 , N5 )

As N1, N3 and N4 is similar and we have high number of succeesful restaurents in N1 and N4 , we can expect N3 is also a good location for restaurent with less competition. For this I have not used Restaurent number is normalization as I am not going to ue it.

In the next cell I will use K-means clustering ifor diff K values to get the best K 

```python
#creating a temporary array to score WCSS [ Within Cluster Sum of Square ] score of clustetrs for diff k values
wcss_scores=[]
for kclusters in (range(2,7)):
    # run k-means clustering
    kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_scaler)
    print(kclusters," ",kmeans.inertia_)
    wcss_scores.append([kclusters,kmeans.inertia_])
```

Let me plot the values in a grapgh to find the best k using ELBOW METHOD

```python
pd.DataFrame(wcss_scores,columns=['K - Value','WCSS Score']).plot(kind='line',x='K - Value',y='WCSS Score',grid=True,figsize=(10,6))
```

As we can see , from the plot K=5 is better . Hence we will use K= 5 to make cluster among the neighbourhood 

```python
# applying k means algorithm for k=5
kmeans = KMeans(n_clusters=5, random_state=0).fit(df_scaler)

#Adding the cluter number in the dataframe 
final_consolidated_df.insert(0, 'Cluster Labels', kmeans.labels_)

final_consolidated_df.head()
```
